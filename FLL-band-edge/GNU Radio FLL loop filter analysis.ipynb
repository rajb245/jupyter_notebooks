```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (7, 3.5)
```

```python
freq = np.fromfile('/tmp/freq.f32', 'float32')
phase = np.fromfile('/tmp/phase.f32', 'float32')
error = np.fromfile('/tmp/error.f32', 'float32')
freq_new = np.fromfile('/tmp/freq_new.f32', 'float32')
phase_new = np.fromfile('/tmp/phase_new.f32', 'float32')
error_new = np.fromfile('/tmp/error_new.f32', 'float32')
```

```python
N = 4_000
plt.plot(freq[:N], label='Legacy loop filter')
plt.plot(freq_new[:N], label='Fixed loop filter')
plt.title('FLL frequency accumulator lock-in')
plt.xlabel('Sample number')
plt.ylabel('Frequency (rad/sample)')
plt.legend();
```

```python
plt.plot(freq, label='Legacy loop filter')
plt.plot(freq_new, label='Fixed loop filter')
plt.title('FLL frequency accumulator')
plt.xlabel('Sample number')
plt.ylabel('Frequency (rad/sample)')
plt.legend();
```

```python
plt.plot(np.diff(np.unwrap(phase.astype('float'))), label='Legacy loop filter')
plt.plot(np.diff(np.unwrap(phase_new.astype('float'))), label='Fixed loop filter')
plt.title('FLL frequency (phase derivative)')
plt.xlabel('Sample number')
plt.ylabel('Frequency (rad/sample)')
plt.legend();
```
